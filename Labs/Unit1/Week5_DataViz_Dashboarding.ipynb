### Q1 ‚Äî Exploratory vs. Explanatory
**Prompt:** *"Explain the difference between exploratory and explanatory visualization."*
**Hints:** Consider audience, purpose, level of polish, and how much iteration you expect.

üìù **Your response:**

*   **Exploratory:** Aims to understand data, find patterns, and generate hypotheses. Audience is typically the analyst or team. Visualizations can be rough and iterative. High iteration.
*   **Explanatory:** Aims to communicate specific findings and tell a story to a particular audience. Audience is often stakeholders or the public. Visualizations are highly polished and refined. Lower iteration.

### Q2 ‚Äî Chart Types & When to Use Them
**Prompt:** *"Describe 5 common chart types (e.g., bar, line, scatter, histogram, map) and the type of data analysis question each is best suited to answer."*

For each chart, list: **Question archetype**, **Data shape**, **Pitfalls**, **Alternatives**.

üìù **Your response (template):**
1. **Bar:**  
   - Question archetype: Comparison across categories (e.g., Which carrier had the highest on-time rate?).
   - Data shape: One categorical variable and one quantitative variable.
   - Pitfalls: Misleading scales (starting at non-zero), too many categories making it hard to read.
   - Alternatives: Column chart (vertical bars), sorted bar chart, grouped or stacked bar chart for multiple series.
2. **Line:**  
   - Question archetype: Trends over time (e.g., How has the average arrival delay changed daily/monthly?).
   - Data shape: Time series data (datetime) and one or more quantitative variables.
   - Pitfalls: Using line charts for non-time series data, too many lines making it cluttered.
   - Alternatives: Area chart, sparklines.
3. **Scatter:**  
   - Question archetype: Relationship between two quantitative variables (e.g., Is there a relationship between cost and revenue?).
   - Data shape: Two quantitative variables. Can include a third variable mapped to color or size.
   - Pitfalls: Overplotting with large datasets, misinterpreting correlation as causation.
   - Alternatives: Bubble chart (with size), 2D histogram, hexbin plot.
4. **Histogram:**  
   - Question archetype: Distribution of a single quantitative variable (e.g., What is the distribution of arrival delays?).
   - Data shape: One quantitative variable.
   - Pitfalls: Using inappropriate bin sizes that distort the shape of the distribution.
   - Alternatives: Density plot, box plot (for comparing distributions across categories).
5. **Map:**  
   - Question archetype: Geographic patterns or distribution (e.g., Which origin cities have the most flights?).
   - Data shape: Geographic data (latitude/longitude, city, state, country) and a quantitative or categorical variable.
   - Pitfalls: Choosing an inappropriate map type, data normalization issues, overplotting points.
   - Alternatives: Choropleth map, symbol map, flow map (for movement between locations).

### Q3 ‚Äî Airline Executive Dashboard KPIs

üìù **Your response (template):**
1. KPI: **On-Time Rate**
   - Why it matters: Indicates operational efficiency and customer satisfaction.

   - Drill-down fields: Carrier, Origin, Destination, Date/Month.

   - Why it matters: Measures how full flights are, indicating revenue potential and asset utilization.

### Create a Synthetic Airline Dataset (if you don't have one)
Use this dataset to prototype visualizations and dashboards. It contains realistic columns for the KPIs you defined. You can always swap in your BigQuery data later.


````python
# Generate a synthetic airline dataset
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(467)

dates = pd.date_range(start='2024-01-01', end='2024-06-30', freq='D')
carriers = ['AA', 'DL', 'UA', 'WN', 'AS']
origins = ['ATL','DFW','DEN','ORD','LAX','SEA','SFO','PHX','CLT','MCO']
dests =   ['JFK','LGA','BOS','SFO','LAX','SEA','MIA','IAH','IAD','PHL']

rows = []
for d in dates:
    for carrier in carriers:
        # number of flights per day per carrier
        n = np.random.randint(80, 140)
        for _ in range(n):
            o = np.random.choice(origins)
            de = np.random.choice(dests)
            if o == de:
                continue
            distance = np.random.randint(300, 2800)
            seats = np.random.choice([90,120,150,180,210])
            load_factor = np.clip(np.random.normal(0.82, 0.07), 0.55, 0.98)
            pax = int(seats * load_factor)
            dep_delay = int(np.random.normal(8, 15))
            arr_delay = dep_delay + int(np.random.normal(0, 12))
            dep_delay = max(dep_delay, -5)
            arr_delay = max(arr_delay, -10)
            cancelled = np.random.rand() < 0.025
            diverted = np.random.rand() < 0.005
            revenue = pax * np.random.uniform(120, 420)
            casm = np.random.uniform(0.08, 0.16)  # cost per available seat mile
            cost = casm * seats * distance
            fuel_burn = distance * np.random.uniform(3.0, 4.5)  # gallons per seat-mile proxy
            nps = np.random.normal(35, 15)  # if you want a customer metric proxy
            rows.append({
                'date': d,
                'carrier': carrier,
                'origin': o,
                'dest': de,
                'distance_miles': distance,
                'seats': seats,
                'passengers': pax,
                'dep_delay_min': dep_delay,
                'arr_delay_min': arr_delay,
                'cancelled': int(cancelled),
                'diverted': int(diverted),
                'revenue_usd': round(revenue, 2),
                'cost_usd': round(cost, 2),
                'fuel_burn_gal': round(fuel_burn, 1),
                'nps': round(nps, 1)
            })

df_flights = pd.DataFrame(rows)

# Derived KPIs
df_flights['on_time'] = (df_flights['arr_delay_min'] <= 15).astype(int)
df_flights['load_factor'] = (df_flights['passengers'] / df_flights['seats']).round(3)
df_flights['yield_usd_per_pax_mile'] = (df_flights['revenue_usd'] / (df_flights['passengers'] * df_flights['distance_miles']).clip(lower=1)).round(4)
df_flights['asm'] = df_flights['seats'] * df_flights['distance_miles']  # available seat miles
df_flights['casm'] = (df_flights['cost_usd'] / df_flights['asm']).round(4)
df_flights['month'] = df_flights['date'].dt.to_period('M').astype(str)

df_flights.head()
````

#### Prompt E ‚Äî Cost vs. Revenue with CASM Coloring (Scatter)

````python
# Prompt A

# Calculate on-time rate by carrier
ontime_rate_by_carrier = df_flights.groupby('carrier')['on_time'].mean().sort_values(ascending=False) * 100

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=ontime_rate_by_carrier.index, y=ontime_rate_by_carrier.values, palette='viridis')

# Add percentage labels on bars
for i, rate in enumerate(ontime_rate_by_carrier.values):
    plt.text(i, rate + 0.5, f'{rate:.1f}%', ha='center')

plt.title('On-Time Rate by Carrier')
plt.xlabel('Carrier')
plt.ylabel('On-Time Rate (%)')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````
